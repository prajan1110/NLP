# Problem 1: Match the word "monkey" exactly.
# Test String: `The monkey ate the banana.`

```python
text = "The monkey ate the banana."

print(bool(re.search(r"\bmonkey\b", text)))
```

# Problem 2: Match "cat", "cut", and "cot" using the dot (.) wildcard.
# Test String: `The cat cut the cot.`

```python
import re

text = "The cat cut the cot."
print(re.findall(r"c.t", text))
```


# Problem 3: Match "bag", "beg", "big", "bog", "bug" using a character class [].
# Test String: `The big bug sat on the bog with a bag to beg.`


```python
import re
text = "The big bug sat on the bog with a bag to beg."
print(re.findall(r"b[aieou]g", text))
```

# Problem 4: Match the years 2024, 2025, and 2026 using a range [].
# Test String: `In 2024, 2025, and 2026.`

```python
import re
text = "In 2024, 2025, and 2026."
print(re.findall(r"202[2-6]", text))
```
